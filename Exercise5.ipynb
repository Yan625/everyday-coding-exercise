**Exercise 12 Sort by column:**

Problem Statemnt:
    
    Sort a given 2D array of shape (4, 5) by 2nd column (i.e. column at index 1) in ascending order.
    That means, we should re-arrange complete row based on 2nd columns' values.
    Given 2D array is:
         [[21 20 19 18 17]
          [16 15 14 13 12]
          [11 10  9  8  7]
          [ 6  5  4  3  2]]
    Print the 2D array in sorted order.
    Note you have to generate the 2D Array.Here 2nd column means column present at index 1.

Output Format :
     
    [[ 6  5  4  3  2]
    [11 10  9  8  7]
    [16 15 14 13 12]
    [21 20 19 18 17]]

```python
a=[[21, 20, 19, 18, 17],[16, 15, 14, 13, 12],[11, 10,  9,  8,  7],[ 6,  5,  4,  3,  2]]
b=np.array(a)
c=b[:,1]#select out second column

#sort number of second colunm according to the big or small
d=[]
for i in c:
    d.append(i)
d.sort()
d

#find the corresponding index
e=[]
for i in d:
    for j in range(4):
        if b[j][1]==i:
            e.append(j)

#according to the index, reorder the array
f=[]
for index in e:
    f.append(b[index,:])

g=np.array(f)
print(g)  
```

**Exercise 13 Killed in USA:**

Given file "terrorismData.csv"
It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement :

    Find value of killed column only where country == ‘United States’?
    Print 0 in place of missing values.
    Print count of Killed as integer value.

Output Format :

    Killed1
    Killed2
    Killed3
    Killed4
    . . .


```python
import csv
import numpy as np
data1=open('terrorismData.csv')
data=csv.DictReader(data1,skipinitialspace=True)

killed=[]
country=[]
for row in data:
    killed.append(row['Killed'])
    country.append(row['Country'])
    
np_killed=np.array(killed)
np_country=np.array(country)

np_killed[np_killed=='']='0.0'
np_killed=np.array(np_killed,dtype=float)
np_killed=np.array(np_killed,dtype=int)

bool_arr=np_country=='United States'
ans=np_killed[bool_arr]

for i in ans:
    print(i)    
```

**Exercise 14 Total Killed in USA:**

Given file "terrorismData.csv"
It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement :

    Find total number of people killed from USA?
    Note: Some columns of killed are empty so replace them by 0.
    Print count of Killed as integer value.

Output Format :

    TotalKilled

```python
import csv
import numpy as np
data1=open('terrorismData.csv')
data=csv.DictReader(data1,skipinitialspace=True)

killed=[]
country=[]
for row in data:
    killed.append(row['Killed'])
    country.append(row['Country'])
    
np_killed=np.array(killed)
np_country=np.array(country)

np_killed[np_killed=='']='0.0'
np_killed=np.array(np_killed,dtype=float)
np_killed=np.array(np_killed,dtype=int)

bool_arr=np_country=='United States'
ans=np_killed[bool_arr]
print(np.sum(ans)) 
```
