**Exercise 4**

Print name of all online courses offered by Coding Ninjas. You can fetch the data using their open API :
https://codingninjas.in/api/v3/courses.

```python
response1=requests.get('https://codingninjas.in/api/v3/courses')
data1=response1.json()
for i in data1['data']['courses']:
    if i["available_online"] == True:
        print(i['name'])
```

**Exercise 5**

Print name of all the Coding Ninjas events (given in past registered events) which have word “Webinar” in their name.

Data can be fetched from this URL : https://codingninjas.in/api/v3/events


```python
response2=requests.get("https://codingninjas.in/api/v3/events")
data2=response2.json()
for i in data2['data']['past_registered_events']:  
    a="Webinar"
    if a in i['name']:
        print(i['name'])
```

**Exercise 6**

Print the title of course with id=19.

**Exercise 7.Planet with "and"**

Count the total number of cities of country AU from this URL : https://api.openaq.org/v1/cities/

**Exercise 8.Cities from BE**

Print first 5 cities from country “BE”. Fetch the result from given URL.

Order the results by city.

URL - https://api.openaq.org/v1/cities

Documentation - https://docs.openaq.org/

```python
response6=requests.get('https://api.openaq.org/v1/cities',params={'country':'BE'})
data=response6.json()
pprint.pprint(data)
```

```python
city=data['results']
for j in city[0:5]:
    print(j['city']) 
```

**Exercise 9.USA WOEID**

Make an API call to - https://www.metaweather.com/api/location/search/

Search for "USA" and print its woeid.

```python
a=requests.get('https://www.metaweather.com/api/location/search/',params={'query':'USA'})
data=a.json()
for i in data:
    print(i['woeid'])
```

Find name of holiday on 2nd October 2019 in country India.
